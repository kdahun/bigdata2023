````python
class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def Personinfo(self):
    return self.name + ":(age:"+ str(self.age)+")"
````

````python
my_info = Person("김다훈",23)
````

````python
my_info.Personinfo()
````

````python
class Student(Person):
  def __init__(self,name,age,grade):
    Person.__init__(self,name,age)
    self.grade=grade
  def GetStudent(self):
    return self.Personinfo() + ", " + str(self.grade )
````

````python
my_info = Student("김다훈",23,3)
my_info.GetStudent()
````

````python
my_info.Personinfo()
````

````python
class MyScore:
  def __init__(self,kor,eng):
    self.kor = kor
    self.eng = eng
  def getTot(self):
    return (self.kor+self.eng)

class MyScore_Sub(MyScore):
  def __init__(self,kor,eng,mat):
    self.mat = mat
    #super().__init__(kor,eng)
    MyScore.__init__(self,kor,eng)
  def getTot(self):
    #return (super().getTot()+self.mat)
    return (MyScore.getTot()+self.mat)


````

````python
a=MyScore(57,57)
a.getTot()
````

# 메모리제이션으로 제귀함수 효율화

````python
def fibo_memo(n):
  d=[0]*100
  d[1]=1
  d[2]=1

  for i in range(3,n+1):
    d[i]=d[i-1]+d[i-2]
  return d[n]

fibo_memo(5)
````

````python
import requests
from bs4 import BeautifulSoup
for j in range(18,23):
  res = requests.get(f"https://search.daum.net/search?w=tot&q=20{j}%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR")
  print(res.status_code)
  print(len(res.text))

  bSoup = BeautifulSoup(res.text,"lxml")
  mov=bSoup.find("ol",attrs={"class":"type_plural list_exact movie_list"})
  mov=mov.find_all('li')

  for i in range(0,5):
    img_url = mov[i].find("img",attrs={"class":"thumb_img"})['src']
    img_res = requests.get(img_url)
    with open(f"20{j}년도 rank{i}.jpg","wb") as f:
      f.write(img_res.content)
````
