```python
def dataframe_from_json_files(dir_name):
  df = pd.DataFrame()
  json_files = os.path.join(dir_name,'*.json')
  print(json_files)
  file_list = glob.glob(json_files)
  print(file_list)
  for path in file_list:
    temp=pd.read_json(path,lines=True)
    df=pd.concat([df,temp],ignore_index=True)
  return df

answer_df = dataframe_from_json_files("실력평가001/1_문항정오답표")
item_df = dataframe_from_json_files("실력평가001/2_문항IRT")
user_df = dataframe_from_json_files("실력평가001/3_응시자IRT")
```

```python
display(answer_df.head())
display(item_df.head())
display(user_df.head())
```

```python
res_order = pd.merge(answer_df,item_df,on = ["testID","assessmentItemID"],how = "right")
res_order
```

```python
res_order = pd.merge(res_order,user_df,on=["learnerID","testID"],how="right")
res_order
```

```python
# 성별 만들기
list_sex = []
list_section = []
list_age=[]

for res in res_order['learnerProfile_x']:
  ssa=res.split(";")
  list_sex.append(ssa[0])
  list_section.append(ssa[1])
  list_age.append(ssa[2])


res_order['sex']=list_sex
res_order['section']=list_section
res_order['age']=list_age

```
