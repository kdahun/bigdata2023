```python
from sklearn import datasets

data = datasets.load_iris()
data
```

```python
import pandas as pd
import numpy as np
iris = pd.DataFrame(data['data'],columns=data['feature_names'])

# 각 열의 평균
# iris.loc['평균']=iris.mean(0)
# display(iris)

# 각 행의 평균
print(iris.iloc[:150,:].mean(1))
iris['평균'] = iris.iloc[:150,:].mean(1)
display(iris)

# target 정보 추가
iris['target']=data['target']
display(iris)

# 평균

print(iris['target']==0)
print(iris['target']==1)
print(iris['target']==2)

iris_mean = iris.iloc[:,:4].mean(0)
iris0_mean = iris[iris['target']==0].iloc[:,:4].mean(0)
iris1_mean = iris[iris['target']==1].iloc[:,:4].mean(0)
iris2_mean = iris[iris['target']==2].iloc[:,:4].mean(0)

print(iris_mean)
print(iris0_mean)
print(iris1_mean)
print(iris2_mean)
```

```python
print(iris.info())
print(data['target'].shape)
```

```python
df = pd.DataFrame()
df['iris_mean']=iris_mean
df['iris0_mean']=iris0_mean
df['iris1_mean']=iris1_mean
df['iris2_mean']=iris2_mean
display(df)
# mean과 다른 것과 같은지 확인

#print(np.round(df.iris_mean(),2)==(np.round((df.iris0_mean + df.iris1_mean + df.iris2_mean)/3,2)))

print(df['iris_mean']==df['iris0_mean'])
print(df['iris_mean']==df['iris1_mean'])
print(df['iris_mean']==df['iris2_mean'])
print(df['iris_mean']==df.iloc[:,1:])


# 전체 평균과 얼마나 차이가 나는 지 계산

display(df.sub(df.iris_mean,axis=0))

df['iris_mean-iris0_mean']=df['iris_mean']-df['iris0_mean']
df['iris_mean-iris1_mean']=df['iris_mean']-df['iris1_mean']
df['iris_mean-iris2_mean']=df['iris_mean']-df['iris2_mean']
display(df)
```

```python
# target_names 넣기

print(data['target_names'])
target_names = data['target_names']
print(iris['target']==0)

iris.loc[iris['target']==0,'target_names']=target_names[0]
iris.loc[iris['target']==1,'target_names']=target_names[1]
iris.loc[iris['target']==2,'target_names']=target_names[2]

#iris['target'][iris['target']==0]=target_names[0]
#iris['target'][iris['target']==1]=target_names[1]
#iris['target'][iris['target']==2]=target_names[2]

display(iris)


iris["specise"]= iris["target"].replace([0,1,2],["A","B","C"])
display(iris)
```

# 데이터 합치기
merge

```python
df1 = pd.DataFrame({"이름":list("ABC"),"성별":["남자","여자","남자"]})
df1
```

```python
df2 = pd.DataFrame({"이름":list("ABD"),"키":[180,160,150]})
df2
```

```python
# 같은 column 이름을 기준으로 합친다.
display(pd.merge(df1,df2)) # default는 inner join이다.
display(pd.merge(df1,df2,how="left"))
display(pd.merge(df1,df2,how="right"))
display(pd.merge(df1,df2,how="outer"))
```

```python
df3=pd.DataFrame({"이름":list("ABC"), "성별":["남자","여자","남자"],
                  "학번":[1,2,3]})
df4=pd.DataFrame({"이름":list("김박최"), "키":["180","160","150"],
                  "학번":[1,2,4]})
print(df3,"\n",df4)
pd.merge(df3,df4 ,on="학번",how="outer")
```

```python
score1=pd.DataFrame({"이름":list("DACBE"), "영어":[90, 80, 70, 65, 100],
              "수학":[100,90,90,65,70], "학번":[5,7,8,4,2]})
score2=pd.DataFrame({"name":list("DACBE"), "국어":[90, 80, 70, 65, 100],
              "id":[5,7,8,4,2]})
display(score1)
display(score2)
```

```python
pd.merge(score1,score2,left_on="학번",right_on="id")
```

### concat : 축따라 이어 붙이기, 여러개 이어붙이기 가능 => 입력은 리스트로

```python
value1=[list("ABCDE"),[180,160,150,170,170],[23,46,33,65,28],["서울","부산","대전","군산","제주"]]
value2=[list("FGHIJ"),[190,180,150,160,170],[22,18,87,45,74],["전주","서울","수원","부산","인천"]]


```

```python
value1=[["A",180,23,"서울"],["B",160,46,"부산"],["C",150,33,"대전"],["D",170,65,"군산"],["E",170,28,"제주"]]
value2=[["F",190,22,"전주"],["G",180,18,"서울"],["H",150,87,"수원"],["I",160,45,"부산"],["J",170,74,"인천"]]

df1 = pd.DataFrame(value1,columns=["이름","키","나이","주소"])
df2 = pd.DataFrame(value2,columns=["이름","키","나이","주소"])

display(df1)
display(df2)
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```
