```python
from sklearn import datasets

data = datasets.load_iris()
data
```

```python
val = data['data']
target = data['target']
target_names = data['target_names']
feature_names = data['feature_names']
```

```python
iris_df = pd.DataFrame(val,columns=feature_names)
iris_df
```

```python
iris_df['target'] = target
iris_df
```

```python
iris_df[iris_df['target']==0]['target']=target_names[0]
iris_df[iris_df['target']==1]['target']=target_names[1]
iris_df[iris_df['target']==2]['target']=target_names[2]
```

```python
iris_df
```

```python
# 1. iris의 feature_names들의 평균 구하기
iris_df.loc['평균']=iris_df.iloc[:,:4].mean(0)
```

```python
iris_df
```

```python
# target 정보 주가하기 + iris의 target 0번인 컬럼별 데이터 평균
print(iris_df[iris_df['target']=='setosa'].mean(1))
print(iris_df[iris_df['target']=='versicolor'].mean(1))
print(iris_df[iris_df['target']=='virginica'].mean(1))

```
