```python
import nltk
nltk.download("gutenberg")
from nltk.corpus import gutenberg
file_names= gutenberg.fileids()
print(file_names)
```

```python
alice_doc=gutenberg.open("carroll-alice.txt").read()
print("number of charater : ",len(alice_doc))
print(alice_doc[:500])
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
token_alice = word_tokenize(alice_doc) # 토큰화 실행
print("number of words : ",len(token_alice))
print(token_alice)
```

```python
# '와 문자길이가 3 이하인거없애기
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer("[\w']{3,}")
reg_tokens_alice = tokenizer.tokenize(alice_doc.lower())
print("numberf of words : ",len(reg_tokens_alice))
print(reg_tokens_alice[:20])
```

```python
import nltk
nltk.download('stopwords')
```

```python
# 불용어 처리
from nltk.corpus import stopwords # 일반적으로 분석 대상이 아닌 단어들
eng_stops = set(stopwords.words("english"))
result_alice=[word for word in reg_tokens_alice if word not in eng_stops]
```

```python
len(result_alice)
```

```python
result_alice[:100]
```

```python
#단어 카운트 사전:
#result_alice에서 단어를 key로 하고
#등장횟수를 value로 하는 단어 등장 횟수 사전형 자료 만들기
alice_word_count={}
for word in result_alice:
  if word not in alice_word_count.keys() :
    alice_word_count[word]=0
  alice_word_count[word]+=1
sort_alice_word_count=sorted(alice_word_count, key=alice_word_count.get, reverse=True)
type(sort_alice_word_count)
sort_alice_word_count[:10]
```
