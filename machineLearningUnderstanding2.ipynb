```python
bank = pd.read_csv("/content/drive/MyDrive/banklist.csv의 사본",parse_dates=[5,6])
bank.info()
```

```python
bank
```

```python
bank["closing_year"] = bank["Closing Date"].dt.year
bank["closing_month"]= bank["Closing Date"].dt.month
bank["closing_quarter"] = bank["Closing Date"].dt.quarter
```

```python
bank["Closing Date"].dt
```

```python
bank.head()
```

```python
# 각 년도별 파산 은행수
close_year = bank.groupby(bank["closing_year"]).size()
fig,ax = plt.subplots()
ax=close_year.plot()
plt.show()
```

```python
# 각 년도와 쿼터별 파산 은행수
closeYQ = bank.groupby([bank["closing_year"],bank["closing_quarter"]]).size().unstack(fill_value = 0)
closeYQ.plot.bar()
```
